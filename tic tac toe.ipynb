```python
firstuser = ''
seconduser = ''
def askinguser():
    global firstuser
    global seconduser
    firstuser =input("hello and welcome to tic tac toe ..... can you tell which symbol you want to use?: ")
    print('the first user choosed: {}'.format(firstuser))
    if firstuser == 'x':
        seconduser ='o'
    elif firstuser == 'o':
        seconduser ='x'
    print('so the second user will play with: {}'.format(seconduser))    
```

```python
def position(phrase):
    x = int(input(phrase))
    return x
```

```python
def checkinggame():
    row1x = 0
    row1o = 0
    row2x = 0
    row2o = 0
    row3x = 0
    row3o = 0
    col1x = 0
    col1o = 0
    col2x = 0
    col2o = 0
    col3x = 0
    col3o = 0
    diagonal1x = 0
    diagonal1o = 0
    diagonal2x = 0
    diagonal2o = 0
    for num in lis:
        if num == 'x':
            row1x += 1 
        elif num == 'o':
            row1o += 2
    for num in liss:
        if num == 'x':
            row2x += 1
        elif num == 'o':
            row2o += 2
    for num in lisss:
        if num == 'x':
            row3x += 1
        elif num == 'o':
            row3o += 2 
    if lis[0] == 'x' and liss[0] == 'x' and lisss[0] == 'x':
        col1x += 1 
    elif lis[0] == 'o' and liss[0] == 'o' and lisss[0] == 'o':     
        col1o += 2  
    elif lis[1] == 'x' and liss[1] == 'x' and lisss[1] == 'x':
        col2x += 1 
    elif lis[1] == 'o' and liss[1] == 'o' and lisss[1] == 'o':
        col2o += 2 
    elif lis[2] == 'x' and liss[2] == 'x' and lisss[2] == 'x':
        col3x += 1
    elif lis[2] == 'o' and liss[2] == 'o' and lisss[2] == 'o':
        col3o += 2 
    elif lis[0] == 'x' and liss[1] == 'x' and lisss [2] == 'x':
        diagonal1x += 1
    elif lis[0] == 'o' and liss[1] == 'o' and lisss [2] == 'o':
        diagonal1o += 2
    elif lis[2] == 'x' and liss[1] == 'x' and lisss[0] == 'x':
        diagonal2x += 1
    elif lis[2] == 'o' and liss[1] == 'o' and lisss[0] == 'o':
        diagonal2o += 2    
        
    if row1x == 3 or row2x == 3 or row3x == 3:
        return 'x'
    elif row1o == 6 or row2o == 6 or row3o == 6:
        return 'o'
    elif col1x == 1 or col2x == 1 or col3x == 1:
        return 'x'
    elif col1o == 2 or col2o == 2 or col3o == 2:
        return 'o' 
    elif diagonal1x == 1 or diagonal2x == 1:
        return 'x'
    elif diagonal1o == 2 or diagonal2o == 2:
        return 'o'
    else:
        return 'tie'
```

```python
z = 0
v = 0
def game():  
    askinguser()  
    i = 0   
    while i < 9:
        z = position('position of first user')
        board(z,firstuser)
        if checkinggame() == 'x':
            print('X wins !!')
            r = input("do you want to replay? ")
            if r == 'yes':
                i = i - i
                board(0,'')
                continue
            else:
                break
        v =position('position of second user')
        board(v,seconduser)
        if checkinggame() == 'o':
            print('O wins !!')
            r = input("do you want to replay? ")
            if r == 'yes':
                i = i - i
                board(0,'')
                continue
            else:
                break
        i = i + 1
    if i == 9:
        if checkinggame() == 'tie':
            print("so sorry it's a tie ")
            r = input("do you want to replay? ")
            if r == 'yes':
                board(0,'')
                game()
            else:
                pass              
```
